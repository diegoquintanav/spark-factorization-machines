Generamos una lista con los ids de usuarios, items y estilos. Utilizarmeos esta lista para transformar cada id al equivalente al índice dentro de esta, así compactaremos los ids utilizados.

```python
users = list(pd.concat([train, test]).userID.unique())
items = list(pd.concat([train, test]).itemID.unique())
styles = list(pd.concat([train, test]).styleID.unique())
```

```python
users
```

Definimos los offset que tendrán los ids de items y de estilos.

```python
offset_users = 1
offset_items = len(users) + 1 
offset_styles = len(users) + len(items) + 1
```

Aplicamos el cambio de índice a todos los id tanto de el set de entrenamiento como el de test.

```python
train.userID = train.userID.apply(lambda x: users.index(x) + offset_users)
train.itemID = train.itemID.apply(lambda x: items.index(x) + offset_items)
train.styleID = train.styleID.apply(lambda x: styles.index(x) + offset_styles)

test.userID = test.userID.apply(lambda x: users.index(x) + offset_users)
test.itemID = test.itemID.apply(lambda x: items.index(x) + offset_items)
test.styleID = test.styleID.apply(lambda x: styles.index(x) + offset_styles)
```

#### Exportar archivo

Creamos una función que permita exportar el dataframe de pandas en el formato adecuado.

```python
def save_as_file(df, output_file, with_style=False):
    with open(output_file, 'w') as out:
        for idx, row in df.iterrows():
            r = '+1' if float(row['rating']) > 3 else '-1'
            u = str(int(row['userID']))
            i = str(int(row['itemID']))
            
            if with_style:
                s = str(int(row['styleID']))
                out.write("{0} {1}:1 {2}:1 {3}:1\n".format(r,u,i,s))
            else:
                out.write("{0} {1}:1 {2}:1\n".format(r,u,i))
```

Exportamos tanto el set de entrenamiento como el de test en el formato correcto para utilizar máquinas de factorización en estos.

```python
save_as_file(train, output_file='beer-dataset/preprocessed/beers.train')
save_as_file(test, output_file='beer-dataset/preprocessed/beers.test')
```

Guardamos también una con los estilos de las cervezas.

```python
save_as_file(train, output_file='beer-dataset/preprocessed/beers_with_style.train', with_style=True)
save_as_file(test, output_file='beer-dataset/preprocessed/beers_with_style.test', with_style=True)
```
